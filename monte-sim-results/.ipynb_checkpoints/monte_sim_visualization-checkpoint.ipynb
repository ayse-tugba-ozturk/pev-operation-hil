```python
from scipy.io import loadmat
from os.path import dirname, join as pjoin

import matplotlib.pyplot as plt
import matplotlib
import datetime as dt
import pandas as pd
import numpy as np
import os
import pdb
cwd = os.getcwd()
```

```python
filename = "test_agnalknglanga_angla.mat"
mat_fname = pjoin(cwd, filename)
sim_results = loadmat(mat_fname)["sim_results"][0][0][0][0]
```

```python
sim_results_time = sim_results[9][0]
sim_results_choices = sim_results[6][0]
sim_results_choicesProb = sim_results[7]

# Find indices that have choices 
choice_idx_nonan = np.argwhere(~np.isnan(sim_results_choices))
choice_idx_nonan = choice_idx_nonan.reshape(len(choice_idx_nonan))
# Construct arrays from the choices
sim_results_time = sim_results_time[choice_idx_nonan]
sim_results_choices = sim_results_choices[choice_idx_nonan].astype('int')
sim_results_choicesProb_all = sim_results_choicesProb[choice_idx_nonan]
sim_results_choicesProb = sim_results_choicesProb[choice_idx_nonan][np.arange(len(sim_results_choicesProb[choice_idx_nonan])), 
                                                                    sim_results_choices]
# Construct markers for graphing
# TODO: what do 0,1,2 stand for?
marker_dict = {"0": "x", "1": "P", "2":"o"}

numb_event = len(choice_idx_nonan)
```

```python
# Visualization over time horizon
font = {'family' : 'serif',
        'serif' : 'Times New Roman',
        'size'   : 36}
matplotlib.rc('text', usetex=False)
matplotlib.rc('font', **font)

# set mathtext back to regular text
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Times New Roman'
matplotlib.rcParams['mathtext.it'] = 'Times New Roman:italic'
matplotlib.rcParams['mathtext.bf'] = 'Times New Roman:bold'

fig, ax1 = plt.subplots(figsize=(20,8))

choice0 = False
choice1 = False
choice2 = False

s = 100

for act_idx in range(numb_event):
    if (not choice0) and (sim_results_choices[act_idx] == 0):
        plt.scatter(sim_results_time[act_idx], sim_results_choicesProb[act_idx], 
                    marker=marker_dict[str(sim_results_choices[act_idx])], s=s, label="Charing Flex", c="k")
        choice0 = True
    elif (not choice1) and (sim_results_choices[act_idx] == 1):
        plt.scatter(sim_results_time[act_idx], sim_results_choicesProb[act_idx], 
                    marker=marker_dict[str(sim_results_choices[act_idx])], s=s, label="Charging ASAP", c="r")
        choice1 = True
    elif (not choice2) and (sim_results_choices[act_idx] == 2):
        plt.scatter(sim_results_time[act_idx], sim_results_choicesProb[act_idx], 
                    marker=marker_dict[str(sim_results_choices[act_idx])], s=s, label="Leaving", c="g")
        choice2 = True
    else:
        if sim_results_choices[act_idx] == 0:
            plt.scatter(sim_results_time[act_idx], sim_results_choicesProb[act_idx], 
                        marker=marker_dict[str(sim_results_choices[act_idx])], s=s, c="k")
        elif sim_results_choices[act_idx] == 1:
            plt.scatter(sim_results_time[act_idx], sim_results_choicesProb[act_idx], 
                        marker=marker_dict[str(sim_results_choices[act_idx])], s=s, c="r")
        elif sim_results_choices[act_idx] == 2:
            plt.scatter(sim_results_time[act_idx], sim_results_choicesProb[act_idx], 
                        marker=marker_dict[str(sim_results_choices[act_idx])], s=s, c="g")
ax1.set_xlabel(r"Time of Day (Hr)")
ax1.set_ylabel(r"Probability")
ax1.set_title("Charging Choices over Time")
# ax1.set_title(r"\textbf{Charging Choices over Time}")
ax1.set_xticks(np.arange(6, 24, 4))
plt.legend(loc="best", fontsize=28)

ax2 = ax1.twinx()
#--- create TOU list
tou = np.zeros(24)
np.put(tou, np.arange(0,9), 0.217)
np.put(tou, np.arange(9,13), 0.244)
np.put(tou, np.arange(13,19), 0.268)
np.put(tou, np.arange(19,22), 0.244)
np.put(tou, np.arange(22,24), 0.217)
#---
ax2.plot(np.arange(6, 24), tou[6:24])
ax2.set_ylabel(r"Time of Use (\$/kWh)")
ax2.set_yticks(np.linspace(0.22, 0.28, 4))

plt.grid(b=True, alpha=0.5)
# plt.savefig("../visualization/vis_chargingChoice_overTime.pdf")
```

```python
# Visualization over event horizon
font = {'family' : 'serif',
        'size'   : 16}
matplotlib.rc('font', **font)

fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)
width = 0.4

choice0 = False
choice1 = False
choice2 = False

event_indices = np.arange(numb_event)
ax.bar(event_indices, np.sum(sim_results_choicesProb_all, axis=1))
ax.bar(event_indices, sim_results_choicesProb_all[:, 1]+sim_results_choicesProb_all[:, 0])
ax.bar(event_indices, sim_results_choicesProb_all[:, 0])

for act_idx in event_indices:
    if (not choice0) and sim_results_choices[act_idx] == 0:
        ax.plot(act_idx, 0.2, color="k", marker=marker_dict[str(sim_results_choices[act_idx])], 
                markersize=12, label="Charing Flex")
        choice0 = True
    elif (not choice1) and sim_results_choices[act_idx] == 1:
        ax.plot(act_idx, 0.6, color="k", marker=marker_dict[str(sim_results_choices[act_idx])], 
                markersize=12, label="Charging ASAP")
        choice1 = True
    elif (not choice2) and sim_results_choices[act_idx] == 2:
        ax.plot(act_idx, 0.95, color="k", marker=marker_dict[str(sim_results_choices[act_idx])],
                markersize=12, label="Leaving")
        choice2 = True
    else:
        if sim_results_choices[act_idx] == 0:
            ax.plot(act_idx, 0.2, color="k", marker=marker_dict[str(sim_results_choices[act_idx])], markersize=12)
        elif sim_results_choices[act_idx] == 1:
            ax.plot(act_idx, 0.6, color="k", marker=marker_dict[str(sim_results_choices[act_idx])], markersize=12)
        elif sim_results_choices[act_idx] == 2:
            ax.plot(act_idx, 0.95, color="k", marker=marker_dict[str(sim_results_choices[act_idx])], markersize=12)
ax.legend(framealpha=0.8, loc='best')
plt.xlabel("Sequency of Event")
plt.ylabel("Probability")
plt.title("Charging Choices over Events")
plt.show()
```
